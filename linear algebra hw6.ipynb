```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
###Define a rectangle area S
```

```python
x = np.linspace(-30 , 30 , 10)
y = np.linspace(-60 , 60 , 20)
```

```python
###Plotting S
```

```python
xx, yy = np.meshgrid(x , y)
xx = xx.flatten()
yy = yy.flatten()
S = np.vstack( [xx , yy] )
plt.scatter(xx , yy ,color = 'red')
plt.title("S")
plt.xlim(-90 ,90)
plt.ylim(-90, 90)
```

```python
###Liner transformation(rotate pi/3 counterclockwise)
```

```python
angle = np.pi/3
R = np.array([[np.cos(angle),-1*np.sin(angle)],
                [np.sin(angle), np.cos(angle)]])

def rotate(x):
    assert R.shape[1]==x.shape[0], 'incompatible dimension'
    return R@x

X,Y = rotate(S)

plt.scatter(X, Y, color = 'blue')
plt.title("After rotating")
plt.xlim(-90 ,90)
plt.ylim(-90, 90)
```

```python
###Compare the original S and the S after rotating
```

```python
plt.scatter(xx, yy, alpha=0.5)
plt.scatter(X, Y, c='r')
plt.title("Compare")

plt.xlim(-90 ,90)
plt.ylim(-90, 90)
```
